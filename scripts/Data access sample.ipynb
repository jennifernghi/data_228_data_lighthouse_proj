```python
import numpy as np
import pandas as pd

class data_object:
    def __init__(self, _values, _days, _lats, _lons, _col_name, _year):
        self.values = _values
        self.days = _days
        self.lats = _lats
        self.lons = _lons
        self.col_name = _col_name
        self.year = _year
        
        
file = "max_air_temperature_K.npz"
data =  np.load(file, allow_pickle=True) # load npz file
```
